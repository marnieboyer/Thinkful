### I tried to google the solution to changing the times to minutes or seconds, and I got close, but I ended up borrowing this code from https://www.kaggle.com/drgilermo/predicting-the-final-time-of-a-runner-after-21-1k to change time string to minutes.  


```python
def time_to_min(string):
    if string is not '-':
        time_segments = string.split(':')
        hours = int(time_segments[0])
        mins = int(time_segments[1])
        sec = int(time_segments[2])
        time = hours*60 + mins + np.true_divide(sec,60)
        return time
    else:
        return -1
```

```python
time_to_min('1:55:44')
```

## So, I see that the times got slower over the years for both males and females even though the avg age of particpants did not change.  Additionally, in 2016, the 5K average time was actually a little bit faster than 2015.  This made me reflect on my experience with running and the Boston Marathon. Since the marathon is run in April, the weather for the marathon is quite variable.  Warm temps and sun is just as likely as a chilly rain.  Since the marathon is run pretty early in the morning, the weather would impact the finish times more than the earlier times (for example, the first 5K). So, I took a look at Weather Underground data and found the mean, min and max temp for the respective April Mondays on which the Boston Marathon was run in 2015, 16, and 17.

```python
# data to plot
n_groups = 3
meantemp = [46,55,62]
maxtemp = [51,67,75]
mintemp = [41,42,49]

# create plot
#fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.25
opacity = 0.65
 
rects1 = plt.bar(index, meantemp, bar_width,
                 alpha=opacity,
                 color='orange',
                 label='Average Temp')
 
rects2 = plt.bar(index + bar_width, maxtemp, bar_width,
                 alpha=opacity,
                 color='red',
                 label='Max Temp')

rects3 = plt.bar(index + bar_width*2, mintemp, bar_width,
                 alpha=opacity,
                 color='blue',
                 label='Min Temp')
 
plt.xlabel('Year')
plt.ylabel('Temp (F)')
plt.title('Weather Conditions by year')
plt.xticks(index + bar_width , ('2015', '2016', '2017'))
plt.legend()
 
plt.tight_layout()
plt.show()
```

### Interesting, the average and max temps were definitely warmer between 2015 and 2016 and also from 2016 to 2017 which could have impacted finish times year-over-year.  Min temps were all about the same, which may contribute to the closer 5K times.
