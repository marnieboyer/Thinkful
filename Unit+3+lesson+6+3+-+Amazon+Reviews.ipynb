```python
beauty = pd.DataFrame.from_dict(reviews_raw, orient='columns')
```

```python
beauty.head()
```

```python
beauty["alltext"] = beauty["reviewText"].map(str) + beauty["summary"]
beauty.apply(lambda x: x.astype(str).str.lower())
```

```python
keywords = ['good','like','best','easy','nice','great','perfect','recommend', 'easier','love','smells good','A+','excellent','useless','give up','snag',
            'Beautiful',' satisfied','must have', 'happy','brilliant','well made','adorable','excellent','fantastic','amazing','junk','bad','hate'
           ,'priced right','quickly','comfort','sturdy','seamless','helpful','superb','tremendous','best','bargain','avoid','complaint','garbage','disappointing',
            'waste','fool','do not buy','poor','disappoint',' impressed','keen','ideal','winner','favorite','no problem','worse','burn','rash','disappointed'
            'works well','happy','Incredible','wonderful','not good','breaks','cheap','terrible','not','better','nasty','chemical','horrible','horribly',
           'frustrating','disaster','fail','bad']


for key in keywords:
    beauty[str(key)] = beauty.alltext.str.contains(
        str(key),
        case=False
    )
```

```python
beauty['positive'] = np.where((beauty['overall'] >=4),1,0)
```

```python
target = beauty['positive']
```

```python
# Our data is binary / boolean, so we're importing the Bernoulli classifier.
from sklearn.naive_bayes import BernoulliNB

# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()
# Fit our model to the data.
data = beauty[keywords]
bnb.fit(data, target)
# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)
# Display our results.
print("Number of mislabeled points out of a total {} points : {}".format(
    data.shape[0],
    (target != y_pred).sum()
))
print(1-(target != y_pred).sum()/data.shape[0])
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(target, y_pred)
```
