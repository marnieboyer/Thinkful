### I tried to google the solution to changing the times to minutes or seconds, and I got close, but I ended up borrowing this code from https://www.kaggle.com/drgilermo/predicting-the-final-time-of-a-runner-after-21-1k to change time string to minutes.  


```python
def time_to_min(string):
    if string is not '-':
        time_segments = string.split(':')
        hours = int(time_segments[0])
        mins = int(time_segments[1])
        sec = int(time_segments[2])
        time = hours*60 + mins + np.true_divide(sec,60)
        return time
    else:
        return -1
```

## I found Seaborn and I wanted to try it. I wanted to make a grid with the runners' ages. But, since there are people from teenagers to octogenarians, I decided to group the ages - less than 20, 20-29, 30-39, 40-49, 50-59 and 60+. I wrote this function using "floor" rounding to put ages into the right bucket.

```python
## Create a function to put ages into decade groups (0-9, 10-19, 20-29, etc)
import math

def decade(rage):
    if rage <= 19:
        s3 = '19 & under'
    elif rage >= 60:
        s3 = '60 & over'
    else:
        s1 = rage/10
        s2 = math.floor(s1)
        s3 = s2 * 10
    return str(s3)

print(decade(80))         #test my function
```

```python
##Apply round function to my dfs

df15['decade'] = df15['age'].apply(lambda x: decade(x))
df16['decade'] = df16['age'].apply(lambda x: decade(x))
df17['decade'] = df17['age'].apply(lambda x: decade(x))
```
